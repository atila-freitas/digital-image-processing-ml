```python
import cv2

from matplotlib import pyplot as plt

#1 - Extrair a componente "RED" da figura flower
#2 - Aplicar uma função para binarizar a componente RED: parâmetros de entrada (imagem RED e threshold)
#3 - Aplicar uma função para detectar regiões de conexão na imagem de entrada: parâmetros (imagem RED e coordenada de referência)
#4 - Detectar dimensões na região identificada na função anterior utilizando as diferentes medidas de distância: De, D4, D8
```

```python
img = cv2.imread('flower.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
blue, green, red = cv2.split(img)
```

```python
def binarizar_image(image, threshold):
    image_return = image.copy()
    rows, columns = image_return.shape
    for i in range(rows):
        for j in range(columns):
            image_return[i][j] = 0 if image_return[i][j] < threshold else 255
    return image_return
            
```

```python
binary_image = binarizar_image(red, 50)
        
```

```python
plt.imshow(cv2.cvtColor(binary_image, cv2.COLOR_BGR2RGB))
```
